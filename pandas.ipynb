```python
import datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pylab import *

%matplotlib inline
```

```python
filename = "2013_movies.csv"
```

```python
with open(filename, 'rb') as csvf:
    d_f = pd.read_csv(filename)
```

```python
print d_f.dtypes
```

```python
d_f["ReleaseDate"] = pd.to_datetime(d_f["ReleaseDate"])
```

```python
print d_f.dtypes
```

```python
d_f.plot(x="ReleaseDate", y="DomesticTotalGross")
locs,labels = yticks()
yticks(locs, map(lambda x: "%.0f" % x, locs*1e-6))
plt.xlabel("Release Date")
plt.ylabel("Domestic Total Gross (millions)")
plt.title("Domestic Total Gross vs Release Date")
plt.show()
```

```python
d_f.plot(x="DomesticTotalGross", y="Runtime" )
locs,labels = xticks()
xticks(locs, map(lambda x: "%.1f" % x, locs*1e-6))
plt.xlabel("Domestic Total Gross (millions)")
plt.ylabel("Runtime (min)")
plt.title("Runtime vs Domestic Total Gross")
plt.show()
```

```python
df = d_f.dropna(subset=["ReleaseDate"])
```

```python
df['Months'] = df['ReleaseDate'].map(lambda x: x.month)
```

```python
grouped = df.groupby('Months')
```

```python
grouped.groups
```

```python
means = grouped.DomesticTotalGross.mean()
errors = grouped.DomesticTotalGross.std()
```

```python
means
```

```python
errors
```

```python
fig, ax = plt.subplots()
plt.ylabel("Domestic Total Gross (millions)")
plt.title("Domestic Total Gross vs Release Date")
means.plot(x='Months', yerr=errors, ax=ax, kind='bar', figsize=(10,7))
```
